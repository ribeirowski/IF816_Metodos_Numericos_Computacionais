1) Importing libraries:

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

2) Solving the First Equation numerically: dy/dt = (y + 4).cos(t)

*Considering y0 = 1 and initial time = 0

```python
def numeric1 (y, t):
  dydt = (y + 4) * np.cos(t)
  return dydt

y0 = 1
time1 = np.linspace(0, 50, 1000)
numerica1 = odeint(numeric1, y0, time1)
```

2.1) Solving the First Equation analytically: y = (y0 + 4).cos(t)

*Considering constant k = 5 (solving the equation to y0 = 1) and initial time = 0

```python
def analytics1 (k, t):
  yt = (k * np.exp(np.sin(t))) - 4
  return yt

k1 = 5
analitica1 = analytics1(k1, time1)
```

3) Plotting graph of the first equation:

```python
# Plot the data
fig = plt.figure(facecolor='w', figsize = (9, 5))
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.set_title('Questão 6: dy/dt = (y + 4).cos(t)',fontsize = 12)
ax.plot(time1, numerica1, 'r', alpha=0.5, lw=2, ls="-", label='Solução Numérica')
ax.plot(time1, analitica1, 'b', alpha=0.8, lw=2, ls=":", label='Solução Analítica')
ax.set_xlabel('t')
ax.set_ylabel('y(t)')
ax.set_ylim(-4, 13)
ax.grid()
legend = ax.legend()
plt.show()
```

4) Solving the Second Equation numerically: dy/dt = y * exp(t)

*Considering y0 = 1 and initial time = 0

```python
def numeric2 (y, t):
  dydt = y*np.exp(t)
  return dydt

time2 = np.linspace(0, 1, 1000)
numerica2 = odeint(numeric2, y0, time2)
```

4.1) Solving the Second Equation analytically: y = k * exp(exp(t))

*Considering constant k = 1/exp(1) (solving the equation to y0 = 1) and initial time = 0

```python
def analytics2 (k, t):
  yt = k * np.exp(np.exp(t))
  return yt

k2 = 1/np.exp(1)
analitica2 = analytics2(k2, time2)
```

5) Plotting graph of the second equation:

```python
# Plot the data
fig = plt.figure(facecolor='w', figsize = (9, 5))
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.set_title('Questão 6: dy/dt = y.exp(t)',fontsize = 12)
ax.plot(time2, numerica2, 'r', alpha=0.5, lw=2, ls="-", label='Solução Numérica')
ax.plot(time2, analitica2, 'b', alpha=0.8, lw=2, ls=":", label='Solução Analítica')
ax.set_xlabel('t')
ax.set_ylabel('y(t)')
ax.grid()
legend = ax.legend()
plt.show()
```

**6) Conclusion:** we observe that the numerical and analytical equations are graphically equivalent in both cases, proving our calculations
