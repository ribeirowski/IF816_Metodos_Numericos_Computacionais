1) Importing libraries:

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

2) Solving the Equation numericaly: y'' + y = cos(t)

*Considering y(0) = y'(0) = 0 and initial time = 0

```python
def numeric(P, t):
    dPdt = [P[1], np.cos(t) - P[1]]
    return dPdt

y0 = [0, 0]
time1 = np.linspace(0, 10, 1000)
numerica = odeint(numeric, y0, time1)
```

3) Plotting the graph of the numeric solution:

```python
# Plot the data
fig = plt.figure(facecolor='w', figsize = (8, 5))
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.set_title('Solução Numérica y´´ + y = cos(t) com y = y´ = 0', fontsize = 12)
ax.plot(time1, numerica[:,0], 'b', alpha=0.8, lw=2, ls="-", label='Solução Numérica')
ax.set_xlabel('t')
ax.set_ylabel('y(t)')
ax.set_ylim(-4, 6)
ax.grid()
ax.legend()
plt.show()
```

4) Plotting the graph of the particular solution:

```python
def pSolution(t):
    ps = (t * np.sin(t))/2
    return ps

# Plot the data
fig = plt.figure(facecolor='w', figsize = (8, 5))
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.set_title('Solução Particular y(t) = 1/2 * t * sen(t)', fontsize = 12)
ax.plot(time1, pSolution(time1), 'r', alpha=0.8, lw=2, ls="-", label='Solução Particular')
ax.set_xlabel('t')
ax.set_ylabel('y(t)')
ax.set_ylim(-4, 6)
ax.grid()
ax.legend()
plt.show()
```

5) Plotting the graph of both numeric and particular solution and compating the results:

```python
# Plot the data
fig = plt.figure(facecolor='w', figsize = (8, 5))
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.set_title('Solução Numérica vs Solução Particular', fontsize = 12)
ax.plot(time1, numerica[:,0], 'b', alpha=0.8, lw=2, ls="-", label='Solução Numérica')
ax.plot(time1, pSolution(time1), 'r', alpha=0.8, lw=2, ls="-", label='Solução Particular')
ax.set_xlabel('t')
ax.set_ylabel('y(t)')
ax.set_ylim(-4, 6)
ax.grid()
ax.legend()
plt.show()
```
